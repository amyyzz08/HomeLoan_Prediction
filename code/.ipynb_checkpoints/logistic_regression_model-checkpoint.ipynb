```python
# Import Dependencies
import warnings
warnings.simplefilter('ignore')

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
# Read the csv file
filepath = "../data/home_loan_applications.csv"
df = pd.read_csv(filepath)
df.head(10)
```

# Data Cleaning and Preprocessing

```python
# Combining income columns
sum_column = df["ApplicantIncome"] + df["CoapplicantIncome"]
df["Total_Income"] = sum_column
df.head(10)
```

```python
# Remove unneccesary columns
loan = df.drop(["Loan_ID","Loan_Amount_Term","Education","ApplicantIncome","CoapplicantIncome"], axis = 1)
loan.head()
```

```python
# Drop rows containing n/a
clean_df = loan.apply(pd.to_numeric, errors = "coerce")
clean_df = loan.dropna()

clean_df
```

```python
loan_df = pd.get_dummies(clean_df)
loan_df
```

```python
# Drop dummy columns
clean_df = loan_df.drop(columns=["Gender_Female","Married_No","Self_Employed_No","Loan_Status_N"])
clean_df.head()
```

```python
# Assigning X (data) and y (target)

X = clean_df[["Gender_Male",
           "Married_Yes",
           "Dependents_0",
              "Dependents_1",
              "Dependents_2",
              "Dependents_3+",
           "Self_Employed_Yes",
           "Property_Area_Rural",
              "Property_Area_Semiurban",
              "Property_Area_Urban",
           "LoanAmount",
           "Credit_History",
           "Total_Income"]]
y = clean_df["Loan_Status_Y"].values.reshape(-1,1)
print(X.shape, y.shape)
```

Splitting the Data for training and testing


```python
from sklearn.model_selection import train_test_split

X = pd.get_dummies(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

X_train.head()
```

Creating a Logistic Regression Model - Classification Model


```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
# Accuracy score
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```
